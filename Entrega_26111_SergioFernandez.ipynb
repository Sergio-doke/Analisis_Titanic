```python
def analisis_inicial(df, nombre_df="DataFrame", columna_id=None):

    print(f"===== Análisis de {nombre_df} =====")
    print("\nPrimeras filas del DataFrame:")
    display(df.head())
    nulos = df.isna().sum()
    print("\nCantidad de valores nulos por columna:")
    display(nulos)
    duplicados = df.duplicated().sum()
    print("\nCantidad de filas duplicadas:", duplicados)
    if columna_id and columna_id in df.columns:
        duplicados_id = df[columna_id].duplicated().sum()
        print(f"\nDuplicados en '{columna_id}':", duplicados_id)
        repetidos = df[columna_id].value_counts()
        print(f"\nValores más repetidos en '{columna_id}':")
        display(repetidos.head(5))
    print("\n" + "="*80)
```

### #1 Limpieza de datos

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() según el caso.
# Documentar el proceso y los resultados.
# Creamos copias independientes para no modificar los  dataframes originales se usa en arquitectura Medallion, Bronze
ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()
```

```python
ventas_clean["producto"] = (
    ventas_clean["producto"]
    .astype(str) # → convierte todo a texto; no tiene parámetros adicionales.
    .str.title() # Convierte a Título: "juan pérez" → "Juan Pérez"
    .str.strip() # Elimina espacios al inicio y final
    .str.replace( " +"," ", regex=True)  # reemplaza "uno o más espacios consecutivos" por un solo espacio
)
#ventas_clean["producto"] = ventas_clean["producto"].astype(str).str.title().str.strip().str.replace(" +", " ", regex=True)
ventas_clean["categoria"] = (
    ventas_clean["categoria"]
    .astype(str) #→ convierte todo a texto; no tiene parámetros adicionales.
    .str.title() # Convierte a Título: "juan pérez" → "Juan Pérez"
    .str.strip() # Elimina espacios al inicio y final
    .str.replace(" +", " ", regex=True)  # reemplaza "uno o más espacios consecutivos" por un solo espacio
)
ventas.head(5)
ventas.dtypes
```

```python
ventas_clean["fecha_venta"] = pd.to_datetime(ventas_clean["fecha_venta"], errors="coerce", dayfirst=True)
marketing_clean["fecha_inicio"] = pd.to_datetime(marketing_clean["fecha_inicio"], errors="coerce", dayfirst=True)
marketing_clean["fecha_fin"] = pd.to_datetime(marketing_clean["fecha_fin"], errors="coerce", dayfirst=True)
```

```python
#normalizar el precio, sacandole el $
 # Se usa nuevamente agrupación con () para encadenar métodos y mantener legibilidad
ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)                        # Convierte todo a texto
        .str.replace("$", "", regex=False)  # Elimina el símbolo $
        #   "$" → texto literal a reemplazar
        #   ""  → nuevo valor (vacío)
        #   regex=False → interpreta "$" literalmente, no como expresión regular
        .str.replace(",", "", regex=False)  # Elimina comas de miles 1,000  1000
        .str.strip()                        # Quita espacios sobrantes
    )
ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")
```

```python
ventas_clean.head(5)
#ventas_clean.dtypes
```

```python
ventas_clean["cantidad"] = pd.to_numeric(ventas_clean["cantidad"], errors="coerce").astype("Int64")
```

```python
print(ventas_clean.dtypes)
```

```python

# 3 Completar valores null con 0
ventas_clean = ventas_clean.fillna(0)

#elimino filas que en la columna precio o cantidad tengan NaN, siempre y cuando
#no afecte a estadisticas finales, depende el analisis que realice del dato
ventas_clean = ventas_clean.dropna(subset=["precio", "cantidad"])


#Solo 0 en cantidad y precio con el precio del articulo, cuando ya tenemos el precio convertido en float
ventas_clean["cantidad"] = ventas_clean["cantidad"].fillna(0)
#agrupo por producto, me quedo con el precio, y a ese precio lo transformo en el promedio de cada fila que es x aplico una lambda.
#a la columna precio le coloco en el caso de que la celda este null en una fila, el promedio de los precios del mismo producto
ventas_clean["precio"] = ventas_clean.groupby("producto")["precio"].transform(
    lambda x: x.fillna(x.mean())
)
'''
transform() devuelve un resultado del mismo tamaño que el DataFrame original.

Eso es importante porque:

groupby() separa en grupos
transform() permite volver a colocar el resultado fila por fila
x.fillna(x.mean())

Reemplaza los NaN del grupo por el promedio del grupo.
'''

#verificamos null
print(ventas_clean.isna().sum())
```

```python
ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()
```

### #3 Agregación

```python
# TODO: Resumir las ventas por categoría de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar métricas como suma y promedio.
```

```python
resumen_producto = (

    ventas_clean

    .groupby(
        by="producto",
        dropna=False,
        as_index=False,
        observed=False
    )

    .agg(
        ingreso_total=('total', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('total', 'size')
    )

)

# Mostrar resultados
display(resumen_producto.head())
```

```python
top_productos = resumen_producto.sort_values(
    by="ingreso_total",
    ascending=False
)

display(top_productos.head(10))
```

```python
mas_vendidos = resumen_producto.sort_values(
    by="unidades",
    ascending=False
)

display(mas_vendidos.head(10))
```

```python
# Resumen por categoria
# 3) Agregación por categoría con groupby + agg
resumen_categoria = (
    ventas_clean
    .groupby(
        by="categoria",      # Puede ser string o lista de strings si quisiéramos agrupar por varias columnas.
        dropna=False,    # Mantener grupo NaN (si hay filas sin categoría).
        as_index=False   # Dejar la categoría como columna normal (y no como índice).
        # observed: si cat_col es 'category' y queremos mostrar solo categorías presentes → True.
    )
    .agg(
        ingreso_total=('total', 'sum'),   # Suma total de ingresos por categoría.
        unidades=('cantidad', 'sum'),       # Unidades totales vendidas en la categoría.
        ventas=('total', 'size'),         # Cantidad de registros/filas (ventas) en la categoría.
        precio_promedio=('precio', 'mean')  # Precio promedio observado en la categoría.
        # Otras funciones útiles: 'median','max','min','std','var','nunique'...
    )
    .sort_values(
        by='ingreso_total', # Ordenar por ingreso total
        ascending=False,    # Descendente: mayores arriba
        na_position='last', # NaN al final
        ignore_index=True   # Reindexar desde 0
    )
)

# 4) Ticket promedio por venta = ingreso_total / ventas
resumen_categoria = resumen_categoria.assign(
    ticket_promedio_por_venta = resumen_categoria['ingreso_total'] / resumen_categoria['ventas']
    # assign: crea/reescribe columnas. Alternativa: resumen_cat['ticket_promedio_por_venta'] = ...
)

print("Resumen por categoría (ordenado por ingreso_total):")
display(resumen_categoria.head(20))
```
